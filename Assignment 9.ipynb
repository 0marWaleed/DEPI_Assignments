```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Titanic.csv")
df.head()
```

```python
df.plot(kind='scatter', x="Age", y="Survived", color='blue')

plt.title("Relation of Survival and Age")
plt.xlabel("Age")
plt.ylabel("Survival")

plt.show()
```

```python
explode = [0,0.05,0.05]
df_emb = df['Embarked'].value_counts()
df_emb.plot(kind = 'pie',explode = explode, autopct="%1.1f%%") #Autopct shows percantage of each one on chart
plt.show()
```

```python
df['Age'].plot(kind = 'hist')
plt.xlabel('Age')
```

```python
surv = df['Survived'].value_counts()
surv.plot(kind = 'pie',explode = [0,0.05], autopct="%1.1f%%")
```
